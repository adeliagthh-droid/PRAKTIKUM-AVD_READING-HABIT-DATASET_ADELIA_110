# Business Understanding

## Business Objective

Dataset Reading Habit Dataset

Data ini merupakan sebuah data yang berisi informasi mengenai kebiasaan membaca responden, seperti jumlah buku yang dibaca per tahun, usia, dan jenis kelamin. Disini kita akan membantu untuk memahami, mengevaluasi, serta menganalisis pola kebiasaan membaca berdasarkan karakteristik demografis.
Dalam dataset ini terdapat sejumlah baris dan kolom yang merepresentasikan data responden terkait kebiasaan membaca mereka.
Data ini bertujuan untuk membantu memahami tingkat minat baca masyarakat serta melihat potensi pasar dalam industri buku.

## Assess Situation

Situasi dalam dataset ini berasal dari data survei kebiasaan membaca. Hal ini berarti data yang diperoleh sangat bergantung pada jawaban responden dan kemungkinan terdapat bias atau ketidakseimbangan distribusi data.

Karena dataset ini mencakup berbagai karakteristik responden seperti usia dan jenis kelamin, maka diperlukan analisis yang lebih terstruktur untuk memahami pola kebiasaan membaca.
Selain itu, terdapat kemungkinan adanya missing values atau data yang tidak merata di setiap kategori, sehingga perlu dilakukan pemeriksaan lebih lanjut pada tahap berikutnya.

## Analytic Goals & Project Plan

Dalam dataset ini akan dilakukan:

- Mengidentifikasi distribusi jumlah buku yang dibaca per tahun
- Menganalisis perbandingan rata-rata jumlah buku berdasarkan jenis kelamin
- Menganalisis perbandingan rata-rata jumlah buku berdasarkan kelompok usia
- Menghasilkan visualisasi untuk memperjelas pola kebiasaan membaca
- Menyediakan dasar pengambilan keputusan terkait potensi pasar industri buku
