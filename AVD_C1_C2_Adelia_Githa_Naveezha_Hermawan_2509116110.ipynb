### Outliers

```python
results = []

cols = reading.select_dtypes(include=['float64', 'int64'])

for col in cols:
    q1 = reading[col].quantile(0.25)
    q3 = reading[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = reading[(reading[col] < lower_bound) | (reading[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(reading)) * 100

    results.append({
        'Kolom': col,
        'Persentase Outliers': percent_outliers
    })

results_reading = pd.DataFrame(results)
results_reading.set_index('Kolom', inplace=True)

display(results_reading)
```

## Eksplorasi Data (EDA)

### Comparison/Perbandingan

```python
avg_books_by_education = reading.groupby(reading.iloc[:, 3])[reading.iloc[:, 7].name].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_books_by_education.index,
            y=avg_books_by_education.values)

plt.xticks(rotation=45)
plt.title('Average Number of Books by Education')
plt.show()
```

Kesimpulan:

Tingkat pendidikan berpengaruh terhadap kebiasaan membaca. Semakin tinggi tingkat pendidikan seseorang, cenderung semakin banyak jumlah buku yang dibaca dalam setahun. Temuan ini dapat menjadi insight penting dalam memahami segmentasi pembaca berdasarkan latar belakang pendidikan.

### Relationship/Hubungan

```python
num_cols = reading.select_dtypes(include=['int64', 'float64'])

cols_to_plot = [col for col in num_cols.columns if 'Age' in col or 'book' in col.lower()]

plt.figure(figsize=(8,6))
sns.heatmap(num_cols[cols_to_plot].corr(), annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation between Age and Number of Books Read')
plt.show()
```

Kesimpulan:

Berdasarkan hasil heatmap korelasi antara Age dan Jumlah Buku yang Dibaca selama 12 bulan terakhir, terlihat bahwa nilai korelasinya cukup rendah (mendekati 0).

- Interpretasi:
Korelasi yang rendah menunjukkan bahwa usia responden tidak memiliki hubungan yang kuat dengan jumlah buku yang dibaca.

- Artinya, baik responden muda maupun tua memiliki kebiasaan membaca yang bervariasi, dan usia bukan faktor utama yang memengaruhi intensitas membaca dalam dataset ini.
